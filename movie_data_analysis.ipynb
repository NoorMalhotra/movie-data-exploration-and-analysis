```python

import pandas as pd
import matplotlib.pyplot as plt

# Load dataset (Colab: upload movie.csv first)
df = pd.read_csv('movie.csv')

# Quick preview
df.head()
```

```python
print("Columns:", list(df.columns))
print("\nShape:", df.shape)
print("\nNulls per column:\n", df.isnull().sum())

# Standardize column names (lowercase, strip spaces)
df.columns = [c.strip() for c in df.columns]

# If your columns use different names, map them to the expected names:
# (uncomment and change the right side if needed)
col_map = {
    # 'title':'movie_title',
    # 'actor1':'actor_1_name',
    # 'Actor 1':'actor_1_name'
}
# df.rename(columns=col_map, inplace=True)

# Show first rows after rename
df.head()
```

```python
for col in ['gross','budget','num_voted_users','imdb_score','duration','year']:
    if col in df.columns:
        # remove commas, currency symbols and cast
        df[col] = (df[col]
                   .astype(str)
                   .str.replace('[\$,]', '', regex=True)   # remove $ and commas
                   .replace({'nan':'', '': pd.NA})
                  )
        df[col] = pd.to_numeric(df[col], errors='coerce')

# Confirm dtypes
df[['gross','budget','num_voted_users','imdb_score','duration','year']].dtypes
```

```python
if 'genres' in df.columns:
    # common separators: '|' or ',' — try both
    df['genres'] = df['genres'].astype(str).str.replace(',', '|')  # unify separators
    df['genres'] = df['genres'].str.split('|')
    df_exploded = df.explode('genres')
else:
    df_exploded = df.copy()
```
